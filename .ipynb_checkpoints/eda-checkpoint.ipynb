## Max & Min Kuantitas

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Melakukan Seasonal Decomposition
result = seasonal_decompose(df['kuantitas'], model='additive')  # Ganti period sesuai dengan panjang musim

# Menampilkan hasil
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8), sharex=True)

# Plot Observasi
df['kuantitas'].plot(ax=ax1, marker='o', linestyle='-', color='blue')
ax1.set_ylabel('Observasi')
ax1.set_title('Time Series Decomposition')

# Plot Trend
result.trend.plot(ax=ax2, linestyle='-', color='orange')
ax2.set_ylabel('Trend')

# Plot Musiman
result.seasonal.plot(ax=ax3, linestyle='-', color='green')
ax3.set_ylabel('Seasonal')

# Plot Residual tanpa garis
result.resid.plot(ax=ax4, linestyle='', marker='o', color='red')
ax4.axhline(0, color='black', linestyle='--')  # Menambahkan garis pada nilai 0
ax4.set_ylabel('Residual')

plt.xlabel('Tanggal')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['kuantitas'])
plt.show()

plot_pacf(df['kuantitas'])
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['kuantitas'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Kuantitas')
plt.xlabel('kuantitas')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Visualisasi Time Series Data
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['kuantitas'], marker='o', linestyle='-', color='blue')
plt.title('Time Series Data')
plt.xlabel('Tanggal')
plt.ylabel('kuantitas')
plt.grid(True)
plt.show()
```
